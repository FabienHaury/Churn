# Tableau stats

```python
stats = df.groupby("Exited").agg({
    "Age": ['mean', 'median', 'min', 'max'],
    "CreditScore": ['mean', 'std'],
    "Balance": ['mean'],
    "EstimatedSalary": ['mean'],
    "NumOfProducts": ['mean'],
    "CustomerId": 'count'  
})

# Simplifier le MultiIndex des colonnes
stats.columns = ['_'.join(col).strip() for col in stats.columns.values]

# Arrondi des colonnes numériques pertinentes et conversion d'Age en int64
stats['Age_mean'] = stats['Age_mean'].round(0).astype('int64')
stats['Age_median'] = stats['Age_median'].round(0).astype('int64')
stats['CreditScore_mean'] = stats['CreditScore_mean'].round(0).astype('int64')
stats['CreditScore_std'] = stats['CreditScore_std'].round(0).astype('int64')

for col in stats.columns:
    if col not in ['Age_mean', 'Age_median', 'CustomerId_count']:
        stats[col] = stats[col].round(2)

# Calcul du pourcentage d'effectifs par groupe Exited
total_customers = stats['CustomerId_count'].sum()
stats['%_Effectifs'] = (stats['CustomerId_count'] / total_customers) * 100

# Arrondi du pourcentage à 2 décimales
stats['%_Effectifs'] = stats['%_Effectifs'].round(2)

stats.reset_index()
```
